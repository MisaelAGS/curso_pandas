```python
import pandas as pd
import numpy as np

df = pd.read_csv("../data/clientes.csv")
df.head(2)
```

```python
# Criar nova coluna para somar 100 pontos para cada cliente.

df["qtd_pontos_adicionais"] = df["qtdePontos"] + 100
df.head(2)
```

```python
# Criar uma coluna para somar a coluna qtd_pontos_adicionais com a coluna qtdePontos
df["qtdTotalPontos"] = df["qtd_pontos_adicionais"] + df["qtdePontos"]
df.head(2)
```

```python
# Criar coluna para mostrar se o cliente tem email ou twitch (somas as colunas flEmail com a flTwitch)

df["email_ou_twitch"] = df["flEmail"] + df["flTwitch"]
filtro = df["email_ou_twitch"] > 0
df[filtro].head(2)
```

```python
# Criar uma coluna para saber quantas redes sociais cada cliente tem
df["qtd_rede_social"] = df["flEmail"] + df["flTwitch"] + df["flYouTube"] + df["flBlueSky"] + df["flInstagram"]
df.head(4)
```

```python
# Calcular o log da coluna qtdePontos
df["log_qtdePontos"] = np.log(df["qtdePontos"])
df.head(2)
```
