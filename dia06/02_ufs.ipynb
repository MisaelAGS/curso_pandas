```python
import pandas as pd

df = pd.read_html("https://pt.wikipedia.org/wiki/Unidades_federativas_do_Brasil")
uf = df[1]
uf.head()
```

```python
# Método para extrair apenas o ano da coluna "Expectativa de vida (2016)"
def expectativa_vida(valor_completo):
    return valor_completo.split(" ")[0].replace(",", ".")
```

```python
# Aplicar método para toda coluna "Expectativa de vida (2016)"
uf["Expectativa de vida (2016)"] = uf["Expectativa de vida (2016)"].apply(expectativa_vida).astype(float)
uf.head()
```
