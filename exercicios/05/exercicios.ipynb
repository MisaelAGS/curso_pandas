```python
# Exercício 5.5
df = pd.read_csv('../../data/transacoes.csv')
# Ordenar dataframe
df = df.sort_values("dtCriacao")
# Criar coluna data
df["data"] = pd.to_datetime(df["dtCriacao"]).dt.date
# Dropar linhas duplicadas pelas colunas idCliente e data
df = df.drop_duplicates(subset=["idCliente", "data"])
df
```
