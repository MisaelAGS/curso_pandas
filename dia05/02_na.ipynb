```python
import pandas as pd

df = pd.read_csv("../data/clientes.csv")
df
```

```python
# Dropar as linhas que possuem algum n/a
df.dropna()
```

```python
# Dropar apenas as linhas que sejam totalmente n/a
df.dropna(how="all")
```

```python
# Dropar a linha que possui alguma n/a
# Ã‰ o default do mÃ©todo .dropna()
df.dropna(how="any")
```

```python
# Dropar linhas que possuem n/a em alguma coluna especÃ­fica.
df.dropna(how="any", subset=["dtAtualizacao"])
```

```python
# Inserir algum valor em todos os n/a do dataframe
df.fillna(0)
```

```python
# Inserir algum valor no lugar do n/a de alguma coluna
df["dtCriacao"].fillna("2025-01-01 12:00:00.000")
```
