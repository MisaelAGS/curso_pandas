```python
import pandas as pd

df = pd.read_csv("../data/clientes.csv")
df.head()
```

```python
# Criação do método para extrair os últimos caracteres da coluna idCliente
def ultimo_caractere(id_completo):
    return id_completo.split("-")[-1]
```

```python
# Aplicação do método para toda coluna de idCliente

df["novo_id"] = df["idCliente"].apply(ultimo_caractere)
df.head()
```
