```python
import pandas as pd

trans = pd.read_csv('../data/transacoes.csv')
clientes = pd.read_csv('../data/clientes.csv')
clientes.head()
```

```python
trans.head()
```

```python
# trazer as informações de clientes para a tabela de transações
# right é para definir quem será a tabela em que irá buscar as colunas.
# how é o tipo de resultado: left para manter todos os dados da esquerda independente da correspondência. Inner para manter apenas os dados que possuem correspondência.
# on especifica qual é a coluna em comum entre as duas tabelas.
trans.merge(right=clientes, how='left', on=['idCliente'], suffixes=['Trans', 'Clientes']).head(2)
```

```python
# O left_on e o right_on são utilizados quando a coluna em comum entre dois dataframes possuem nomes diferentes
trans.merge(right=clientes, how='left', left_on=['idCliente'], right_on=['idCliente'], suffixes=['Trans', 'Clientes']).head(2)
```
