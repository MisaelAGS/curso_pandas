# Quem teve mais transação de streak?

```python
import pandas as pd
prod = pd.read_csv('../../data/produtos.csv')
prod.head()
```

```python
import pandas as pd
trans_prod = pd.read_csv('../../data/transacao_produto.csv')
trans_prod.head()
```

```python
trans_prod = trans_prod.merge(right=prod, how='left', on=['idProduto'])
trans_prod.head()
```

```python
trans = pd.read_csv('../../data/transacoes.csv')
trans.head()
```

```python
trans = trans.merge(right=trans_prod, how='left', on='idTransacao')
trans.head()
```

```python
filtro = trans['descProduto'] == 'Presença Streak'
trans = trans[filtro]
trans.head()
```

```python
trans = trans.groupby(by='idCliente', as_index=False)['idTransacao'].count()
trans = trans.sort_values(by=['idTransacao'], ascending=False)
trans.head(1)
```
